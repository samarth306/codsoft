# FOR MULTIPLICATION

```python
def multiply(*args):
    result = 1
    for number in args:
        result *= number
    return result
```

# FOR DIVISION

```python
def division(*args):
    if not args:
        return 0
    result = args[0]
    for number in args[1:]:
        if number == 0:
            return "Error: Division by zero"
        result /= number
    return result
```

# FOR ADDITION

```python
def addition(*args):
    result = 0
    for number in args:
        result += number
    return result
```

# FOR SUBSTRACTION

```python
def subtraction(*args):
    if not args:
        return 0
    result = args[0]
    for number in args[1:]:
        result -= number
    return result
```

# MAIN CODE

```python
opr = input('Enter any operation (*, +, -, /): ')
user_input = input('Enter comma-separated numbers: ')
numbers = map(float, user_input.split(','))

if opr == '*':
    result = multiply(*numbers)

elif opr == '+':
    result = addition(*numbers)

elif opr == '-':
    result = subtraction(*numbers)

elif opr == '/':
    result = division(*numbers)

else:
    print('Enter Correct Operation')
    result = None

if result is not None:
    print("Your result is:", result)
```
